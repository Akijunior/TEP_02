```python
import pandas as pd
df = pd.read_csv('train.csv')
# 1) Com base na questão 2 da prova passada, pesquisar e aplicar mais 2
# algoritmos de ML na base da prova e comparar os % de acerto
# 2) Pesquisar e aplicar o mecanismo de "matriz de confusão" explicando
# os resultados das matrizes para os 3 algoritmos envolvidos.
```

```python
x = df[['Sex', 'Embarked', 'Pclass', 'IsAlone']]
y = df['Survived']
x = pd.get_dummies(x)
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1 
df.loc[df['FamilySize'] > 1, 'IsAlone'] = 0
df.loc[df['FamilySize'] == 1, 'IsAlone'] = 1

df[['IsAlone','FamilySize']]
```

```python
from sklearn import linear_model
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.20, random_state=0)

modelo = MultinomialNB()
modelo.fit(x_treino, y_treino)
modelo.predict(x_teste)
modelo.score(x_teste, y_teste)
pd.crosstab(y_teste, modelo.predict(x_teste), rownames=['Real'], colnames=['Predito'], margins=True)
```

```python
linear = linear_model.LinearRegression()

linear.fit(x_treino, y_treino)
linear.predict(x_teste)
linear.score(x_teste, y_teste)
pd.crosstab(y_teste, modelo.predict(x_teste), rownames=['Real'], colnames=['Predito'], margins=True)
#print('Coefficient: \n', linear.coef_)
#print('Intercept: \n', linear.intercept_)
#Predict Output
#predicted= linear.predict(x_teste)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(x_treino, y_treino)
model.score(x_treino, y_treino)
pd.crosstab(y_teste, model.predict(x_teste), rownames=['Real'], colnames=['Predito'], margins=True)
```

```python
from sklearn import tree

model = tree.DecisionTreeClassifier(criterion='gini')
model.fit(x_treino, y_treino)
model.score(x_treino, y_treino)
pd.crosstab(y_teste, model.predict(x_teste), rownames=['Real'], colnames=['Predito'], margins=True)

#Predict Output
#predicted= model.predict(x_teste)
147/179
```

```python
from sklearn.metrics import confusion_matrix
pd.crosstab(y_teste, model.predict(x_teste), rownames=['Real'], colnames=['Predito'], margins=True)

# confusion_matrix(y_teste, model.predict(x_teste))
```

```python
confusion_matrix(y_teste, model.predict(x_teste))
```
